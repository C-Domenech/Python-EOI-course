# Data Science: Matplotlib, Numpy and Pandas

```
from urllib import request
from datascience import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Read two books, fast!


```
huck_finn_url = 'https://www.inferentialthinking.com/data/huck_finn.txt'
huck_finn_file = request.urlopen(huck_finn_url)
huck_finn_text = huck_finn_file.read().decode('utf8')
huck_finn_chapters = huck_finn_text.split('CHAPTER ')[44:]

little_women_url = 'https://www.inferentialthinking.com/data/little_women.txt'
little_women_file = request.urlopen(little_women_url)
little_women_text = little_women_file.read().decode('utf8')
little_women_chapters = little_women_text.split('CHAPTER ')[1:]
```

# Adventures of Huckleberry Finn

Display the chapters of Huckleberry Finn in a table.

```
Table().with_column('Chapters', huck_finn_chapters)
```

Count how many times the names Jim, Tom, and Huck appear in each chapter.


```
counts_df = pd.DataFrame({
    'Chapter': [x for x in np.arange(1, 44, 1)],
    'Jim': [x for x in np.char.count(huck_finn_chapters, 'Jim')],
    'Tom': [x for x in np.char.count(huck_finn_chapters, 'Tom')],
    'Huck': [x for x in np.char.count(huck_finn_chapters, 'Huck')]
})

# counts_df
```

```
ax = plt.gca()
plt.title('Number of times each name appears in every chapter')
counts_df.plot(kind='line', x='Chapter', y='Jim', ax=ax, color='#082562')
counts_df.plot(kind='line', x='Chapter', y='Tom', ax=ax, color='#ec7a0f')
counts_df.plot(kind='line', x='Chapter', y='Huck', ax=ax, color='#2da22c')

plt.show()
```

Cumulative number of times each name appears

```
total_counts_df = pd.DataFrame({
    'Chapter': [x for x in np.arange(1, 44, 1)],
    'Jim': [x for x in np.char.count(huck_finn_chapters, 'Jim').cumsum()],
    'Tom': [x for x in np.char.count(huck_finn_chapters, 'Tom').cumsum()],
    'Huck': [x for x in np.char.count(huck_finn_chapters, 'Huck').cumsum()]
})

# total_counts_df
```

```
ax = plt.gca()
plt.title('Cumulative number of times each name appears')
total_counts_df.plot(kind='line', x='Chapter', y='Jim', ax=ax, color='#082562')
total_counts_df.plot(kind='line', x='Chapter', y='Tom', ax=ax, color='#ec7a0f')
total_counts_df.plot(kind='line', x='Chapter', y='Huck', ax=ax, color='#2da22c')

plt.show()
```
